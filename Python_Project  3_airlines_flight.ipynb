### Q.1. What are the airlines in the dataset, accompanied by their frequencies.

```python
df['airline'].nunique()
```

```python
df['airline'].unique()
```

```python
df['airline'].value_counts()
```

```python
df['airline'].value_counts(ascending=True).plot.barh(color=['lightgreen','lightblue'])
plt.title('Airlines with frequencies')
plt.xlabel('Number of flight')
plt.ylabel('Airline')
plt.show()
```

### Q.2. Show Bar Graphs representing the Departure Time & Arrival Time

```python
X=df['departure_time'].value_counts()
X
```

```python
Y=df['arrival_time'].value_counts()
Y
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
plt.bar(X.index, X.values, color="steelblue")
plt.title("Plot of departure time values", fontsize=14)
plt.xticks(rotation=45, ha="right")   
plt.ylabel("Counts")

plt.subplot(1,2,2)
plt.bar(Y.index, Y.values, color="orange")
plt.title("Plot arrival_time", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Counts")

plt.tight_layout()   
plt.show()
```

### Q.3. Show Bar Graph representing the Source City & Destination City

```python
A=df['source_city'].value_counts()
A
```

```python
B=df['destination_city'].value_counts()
B
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
plt.barh(A.index, A.values, color="Blue")
plt.title("Plot of source_city", fontsize=14)   
plt.xlabel("Counts")

plt.subplot(1,2,2)
plt.barh(B.index, B.values, color="Green")
plt.title("Plot destination_city ", fontsize=14)
plt.xlabel("Counts")

plt.tight_layout()   
plt.show()
```

### Q.4. Does price varies with airlines.

```python
df_airline=df.groupby('airline')['price'].mean()
df_airline
```

```python
sns.catplot(x='airline',y='price',kind='bar',palette='rocket', data=df,hue='class')
plt.show()
```

### Q.5. Does ticket price change based on the departure time and arrival time.

```python
df.groupby('departure_time')['price'].mean()
```

```python
df.groupby('arrival_time')['price'].mean()
```

```python
sns.catplot(x='departure_time',y='price', kind='bar',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.catplot(x='arrival_time',y='price', kind='bar',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.relplot(x='arrival_time', y='price',data=df, col='departure_time',kind='line')
plt.show()
```
